```python
import numpy as np

# Vector example
v = np.array([3, 2])
print("Vector:", v)

# Matrix example
A = np.array([[2, 0], [0, 3]])
print("Matrix A:\n", A)

# Matrix multiplication
result = A @ v
print("A @ v =", result)
```

```python
# Dot product
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Dot product:", dot_product)
print("Also using @ operator:", a @ b)
```

```python
# Matrix transpose
B = np.array([[1, 2], [3, 4], [5, 6]])
print("Original matrix:\n", B)
print("\nTransposed:\n", B.T)
```

```python
# Identity matrix
I = np.eye(3)
print("3x3 Identity matrix:\n", I)
```

```python
# Matrix inverse
C = np.array([[4, 7], [2, 6]])
C_inv = np.linalg.inv(C)
print("Original matrix:\n", C)
print("\nInverse matrix:\n", C_inv)
print("\nVerify (should give identity): \n", C @ C_inv)
```
